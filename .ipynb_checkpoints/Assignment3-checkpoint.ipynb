```python
# 2

import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Load the dataset
df = pd.read_csv('bbb.csv')

# Grouping and adding the total values
category_totals = df.groupby('category')['value'].sum().reset_index()

# Sorting categories by total value for better visualization
category_totals.sort_values(by='value', ascending=False, inplace=True)

# Map colors for categories
color_map = {
    'Bed': '#f9c154',
    'Bath': '#5094c3',
    
}

# Assign colors if specified, otherwise default color
category_colors = [color_map.get(category, '#7c4ea6') for category in category_totals['category']]

# Create the treemap with delineated edges for categories
fig, ax = plt.subplots(figsize=(12, 8))
squarify.plot(sizes=category_totals['value'], label=category_totals['category'], color=category_colors, alpha=1, edgecolor='white', ax=ax)

# Get the positions of the categories
cat_rects = squarify.squarify(category_totals['value'], 0, 0, 100, 100)

# Plot each sub-rectangle
# Iterate over categories to plot items within each category
for rect, (cat, total_value) in zip(cat_rects, category_totals.itertuples(index=False)):
    # Filter the dataframe for the current category and create a copy
    cat_df = df[df['category'] == cat].copy()

    # Sort items by value within the category
    cat_df.sort_values(by='value', ascending=False, inplace=True)

    # Calculate the relative size of each item within the category
    item_values_normalized = cat_df['value'] / total_value
    item_areas = item_values_normalized * rect['dx'] * rect['dy']

    # Get the sub-rectangles for the items within this category
    item_rects = squarify.squarify(item_areas.tolist(), rect['x'], rect['y'], rect['dx'], rect['dy'])

    # Plot each sub-rectangle
    for item_rect in item_rects:
        # Draw the rectangle
        ax.add_patch(
            plt.Rectangle(
                (item_rect['x'], item_rect['y']), 
                item_rect['dx'], 
                item_rect['dy'], 
                facecolor=color_map.get(cat, '#7c4ea6'), 
                edgecolor='black', 
                linewidth=0.5
            )
        )

    # Draw the main rectangle with white edges
    ax.add_patch(
        plt.Rectangle(
            (rect['x'], rect['y']), 
            rect['dx'], 
            rect['dy'], 
            fill=False, 
            edgecolor='white', 
            linewidth=0.5
        )
    )

for text in plt.gca().texts:
    text.set_color('white')
    
    
# Set the title and remove the axes
title = 'Q2: LAYERING HIERARCHY'
plt.text(0.5, -0.05, title, fontsize=15, ha='center', va='center', transform=ax.transAxes)
ax.axis('off')

    
# Show the plot
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import squarify
import textwrap

# Load the dataset
df = pd.read_csv('bbb.csv')

# Define colors for categories
color_map = {
    'Bed': '#f9c154',
    'Bath': '#5094c3',
    'Beyond': '#7c4ea6'  
}

# Define a new 'Beyond' category to include everything except 'Bed' and 'Bath'
df['category_agg'] = df['category'].apply(lambda x: x if x in ['Bed', 'Bath'] else 'Beyond')

# Calculate totals for the new aggregated category
agg_category_totals = df.groupby('category_agg')['value'].sum().reset_index()

# Sort categories by total value for better visualization
agg_category_totals.sort_values(by='value', ascending=False, inplace=True)

# Create the treemap
fig, ax = plt.subplots(figsize=(12, 8))
squarify.plot(sizes=agg_category_totals['value'], label=agg_category_totals['category_agg'],
              color=[color_map.get(cat) for cat in agg_category_totals['category_agg']], alpha=0.7, edgecolor='white', ax=ax)

# Enhance text labels for the main categories with bold font and shadow
for text in ax.texts:
    text.set_fontweight('bold')
    text.set_fontsize(16)
    text.set_color('white')
   

# Get the positions of the main categories
cat_rects = squarify.squarify(agg_category_totals['value'], 0, 0, 100, 100)

# Iterate over the aggregate categories to plot subcategories with white edges
for rect, (index, row) in zip(cat_rects, agg_category_totals.iterrows()):
    cat = row['category_agg']
    if cat == 'Beyond':
        # Get the dataframe for subcategories within 'Beyond'
        beyond_df = df[df['category_agg'] == 'Beyond']
        subcategory_totals = beyond_df.groupby('category')['value'].sum().reset_index()
        
        # Adjust the position and size for subcategory rectangles to ensure white edges are visible
        adjusted_sub_rects = squarify.squarify(subcategory_totals['value'], rect['x'] + 0.001, rect['y'] + 0.001, rect['dx'] - 0.002, rect['dy'] - 0.002)

        for sub_rect, (sub_cat, sub_value) in zip(adjusted_sub_rects, subcategory_totals.itertuples(index=False)):
            x, y, dx, dy = sub_rect['x'], sub_rect['y'], sub_rect['dx'], sub_rect['dy']
            ax.add_patch(plt.Rectangle((x, y), dx, dy, facecolor=color_map['Beyond'], edgecolor='white', linewidth=2))
            wrapped_label = textwrap.fill(sub_cat, width=10)
            ax.text(x + dx / 2, y + dy / 2, wrapped_label, ha='center', va='center', 
                    fontsize=9, color='white')

            # Plot the sub-subcategories (individual items) within each subcategory
            items_df = beyond_df[beyond_df['category'] == sub_cat]
            item_rects = squarify.squarify(items_df['value'], x + 0.001, y + 0.001, dx - 0.002, dy - 0.002)
            for item_rect in item_rects:
                ax.add_patch(plt.Rectangle(
                    (item_rect['x'], item_rect['y']), item_rect['dx'], item_rect['dy'], 
                    facecolor=color_map['Beyond'], edgecolor='black', linewidth=0.5
                ))

title = 'Q3: FINISHING TOUCHES'
plt.text(0.5, -0.05, title, fontsize=15, ha='center', va='center', transform=ax.transAxes)
plt.axis('off')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import circlify

# Load the dataset
df = pd.read_csv('bbb.csv')

# Prepare the data for circlify
data = df.groupby('category')['value'].sum().reset_index()

# Convert the category_totals to a list of dictionaries as required by circlify
data = [{'id': row['category'], 'datum': row['value']} for _, row in data.iterrows()]

# Generate the circles with circlify
circles = circlify.circlify(data, show_enclosure=True)

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 12))

# Remove axes
ax.axis('off')

# Find axis boundaries
lim = max(
    max(abs(circle.x) + circle.r, abs(circle.y) + circle.r)
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

# Define a color map for the categories
color_map = {
    'Bed': '#f9c154',
    'Bath': '#5094c3',
    'default': '#7c4ea6'  # Default color for other categories
}

# Draw circles and labels
for circle in circles:
    # Extract the label from the original data
    label = circle.ex.get('id') if circle.ex else None
    # If the circle has an id and it's in our color map, use the corresponding color, otherwise use the default color
    color = color_map.get(label, color_map['default'])
    # Draw the circle border with the specified color
    ax.add_patch(plt.Circle((circle.x, circle.y), circle.r, edgecolor=color, facecolor='none', linewidth=2))
    # Place the label in the center of the circle
    if label:
        plt.text(circle.x, circle.y, label, ha='center', va='center', fontsize=9, color=color)

# Show the plot
plt.show()
```
