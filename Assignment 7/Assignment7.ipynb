```python
#1 

import geopandas as gpd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import pandas as pd
from datetime import timezone
import numpy as np

# Load the shapefiles
canada = gpd.read_file('canada.shp')
umbra = gpd.read_file('umbra_lo.shp')

# Use the CRS from umbra_lo.shp
canada = canada.to_crs(umbra.crs)

# Calculate Newfoundland Time, which is UTC-3:30 or UTC-2:30 after daylight saving time applied
time_zone_offset = timedelta(hours=2, minutes=30)  
umbra['NewfoundlandTime'] = umbra['UTCSec'].apply(lambda x: (datetime(2024, 4, 8, tzinfo=timezone.utc) + 
                                                             timedelta(seconds=x)) - time_zone_offset)
umbra['TimeRounded'] = umbra['NewfoundlandTime'].dt.round('3min')
umbra_3min_intervals = umbra.groupby(umbra['TimeRounded']).first().reset_index()

# Start plotting
fig, ax = plt.subplots(figsize=(15, 10))


canada.plot(ax=ax, color='#9b9e7f', edgecolor='black', linewidth=1.5)

# Plot the umbra shapes with partial transparency
for idx, row in umbra_3min_intervals.iterrows():
    # Extract the polygon for the umbra shape and plot it
    ax.add_patch(plt.Polygon(np.array(row.geometry.exterior.coords), closed=True, color='black', alpha=0.5))
    # Annotate the plot with Newfoundland Time in HH:MM format
    ax.annotate(text=row['NewfoundlandTime'].strftime('%H:%M'), 
                xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                ha='center', va='center', color='white', fontsize=8)

ax.set_title("The 2024 Total Solar Eclipse", fontsize=13, pad=10)

# Set the bounds to focus on Atlantic Canada
ax.set_xlim(-71, -50)
ax.set_ylim(42, 52)
ax.set_axis_off()

# Show the plot
plt.show()
```

```python
#2

import folium
import geopandas as gpd
import pandas as pd

# Load the shapefiles
umbra = gpd.read_file('umbra_lo.shp')

# Ensure the umbra data is using WGS84 latitude and longitude
umbra = umbra.to_crs(epsg=4326)

# Filter out umbra data for every 3-minute interval
umbra['datetime'] = pd.to_datetime(umbra['UTCTime'], format='%H:%M:%S', errors='ignore')
umbra['rounded_time'] = umbra['datetime'].dt.round('3T')  
umbra_3min = umbra.drop_duplicates(subset=['rounded_time'])

# Generate a simplified DataFrame for faster rendering
simplified_umbra = gpd.GeoDataFrame(umbra_3min, geometry=umbra_3min.geometry.simplify(0.01))

# Initialize Folium map centered on Atlantic Canada with ESRI WorldTopoMap tileset
esri_attribution = 'Tiles &copy; Esri &mdash; Esri, DeLorme, NAVTEQ'
esri_tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Topo_Map/MapServer/tile/{z}/{y}/{x}'

m = folium.Map(location=[47, -62], zoom_start=6, tiles=esri_tiles, attr=esri_attribution)

# Style function for the umbra shapes
style_function = lambda x: {
    'fillColor': 'grey',
    'color': 'grey',
    'weight': 1,
    'fillOpacity': 0.5
}

# Add the umbra shapes to the map
for _, row in simplified_umbra.iterrows():
    folium.GeoJson(
        data=row['geometry'],
        style_function=style_function
    ).add_to(m)

# Display the Folium map
m  
```
