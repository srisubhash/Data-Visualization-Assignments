```python
# 1

import pandas as pd

df = pd.read_csv('plastic-production.csv')

# filtering to ignore rows "Total"
filtered_df = df[df['Applications'] != 'Total']
total_sum_per_year = filtered_df.groupby('Time')['Value'].sum()
total_values = df[df['Applications'] == 'Total'].groupby('Time')['Value'].unique().str[0]
difference=(total_sum_per_year-total_values).abs()
print('Showing the difference between sum of total values calculated and total sum given in the dataset per year.')
print(difference)
```

```python

# 2

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv("plastic-production.csv")

# Extract data for total production and production by various sectors
total_production = df[df['Applications'] == 'Total']
production_by_sectors = df[df['Applications'] != 'Total']

# Plotting
plt.figure(figsize=(10, 6))

# Plot total production with solid black line
plt.plot(total_production['Time'], total_production['Value'], label='Total', color='black')

# Plot production by various sectors with different colors and line styles
sectors_to_label = ['Building & construction', 'Packaging', 'Other']
colors = {'Building & construction': 'blue', 'Packaging': 'purple', 'Other': 'red'}
line_styles = 'dotted'

for sector in production_by_sectors['Applications'].unique():
    color = colors.get(sector, 'grey')
    linestyle = line_styles if sector not in sectors_to_label else '-'
    if sector in sectors_to_label:
        plt.plot(production_by_sectors.loc[production_by_sectors['Applications'] == sector, 'Time'],
                 production_by_sectors.loc[production_by_sectors['Applications'] == sector, 'Value'],
                 label=sector, color=color, linestyle=linestyle)
    else:
        plt.plot(production_by_sectors.loc[production_by_sectors['Applications'] == sector, 'Time'],
                 production_by_sectors.loc[production_by_sectors['Applications'] == sector, 'Value'],
                 color=color, linestyle=linestyle)

# Set plot properties
plt.title('Global Plastic Production')
plt.xlabel('Year')
plt.ylabel('Million Tons')

handles, labels = plt.gca().get_legend_handles_labels()

#specify order of items in legend to recreate exactly
order = [0,3,2,1]

#add legend to plot
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order])
  
#show the plot
plt.show() 
```

```python
# 3

import pandas as pd
import matplotlib.pyplot as plt


# Load the dataset into a pandas DataFrame
df = pd.read_csv("river-plastics.csv")

# Create Figure and Axes objects
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot with log scale for x and y axes
sc = ax.scatter(df['Coast length'] / df['Area'], df['Ratio ME/MPW'],
                c=df['Rainfall'], cmap='viridis', alpha=0.8, edgecolors='black')

# Set log scale for x and y axes
ax.set_xscale('log')
ax.set_yscale('log')

# Add colorbar with label
cbar = fig.colorbar(sc, label='Rainfall (mm/Year)')

# Set axes labels and plot title
ax.set_xlabel('Ratio of Coastal Length to Area')
ax.set_ylabel('Plastic Waste in Oceans(% of total plastic waste)')
ax.set_title('Drivers of Plastic Pollution in the Oceans')

# Show the plot
plt.show()
```
